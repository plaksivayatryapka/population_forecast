```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
import numpy as np

from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
df = pd.read_csv('data/WPP2019_FERT_F04_TOTAL_FERTILITY.csv')
df
```

```python
year_columns = [i for i in df.columns if '-' in i]
year_columns
```

```python
cmap = matplotlib.cm.get_cmap('Spectral')

plt.figure(figsize=[10, 6])

for n, year_column in enumerate(year_columns):
    sns.distplot(df[year_column], label=year_column, color=cmap(n/len(year_columns)), bins=30)

plt.xlim([0, 10])
plt.legend()
plt.xlabel('Fertility')
plt.title("Global Fertility Distributions by Year")
plt.savefig('distr1.png', bbox_inches='tight', dpi=300);
```

```python
cmap = matplotlib.cm.get_cmap('Spectral')

plt.figure(figsize=[10, 6])

for n, year_column in enumerate(year_columns):
    sns.histplot(df[year_column], label=year_column, color=cmap(n/len(year_columns)), bins=30)

plt.xlim([0, 10])
plt.legend()
plt.xlabel('Fertility')
plt.title("Global Fertility Distributions by Year")
plt.savefig('distr3.png', bbox_inches='tight', dpi=300);
```

```python
cmap = matplotlib.cm.get_cmap('Spectral')

plt.figure(figsize=[10, 6])

for n, year_column in enumerate(year_columns):
    sns.kdeplot(df[year_column], label=year_column, color=cmap(n/len(year_columns)))
    
plt.xlim([0, 10])
plt.legend()
plt.xlabel('Fertility')
plt.title("Global Fertility Distributions by Year")
plt.grid(linestyle='--', alpha=0.1)
plt.savefig('distr2.png', bbox_inches='tight', dpi=300);
```

```python
plt.figure(figsize=[16, 9])
for country, group in df.groupby('Region, subregion, country or area *'):
    lst = list()
    for year_column_1, year_column_2 in zip(year_columns, year_columns[1:]):
        value = group[year_column_2].median() / group[year_column_1].median()
        lst.append(value)
    
    plt.plot(year_columns[1:], lst)
    
plt.grid(linestyle='--', alpha=0.2)
plt.title('5-year fertility dynamics');
```

```python
plt.figure(figsize=[10, 6])
for n, (year_column_1, year_column_2) in enumerate(zip(year_columns, year_columns[1:])):
    values = df[year_column_2] / df[year_column_1]

    sns.kdeplot(values, color=cmap(n/len(year_columns)), label=year_column_2)
    
plt.grid(linestyle='--', alpha=0.2)
plt.legend()
plt.xlim([0.8, 1.2])
plt.title('5-year fertility growth distribution')
plt.savefig('distr4.png', bbox_inches='tight', dpi=300);
```
